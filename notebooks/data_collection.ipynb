```python
import yfinance as yf
import pandas as pd
import os

os.makedirs("../data/raw", exist_ok=True)

tickers = {
    "FTSE100": "^FTSE",
    "UK_Gov_Bonds": "ISF.L",
    "GBPUSD": "GBPUSD=X",
    "SP500": "^GSPC"
}

for name, ticker in tickers.items():
    df = yf.download(ticker, start="2015-01-01", end="2025-01-01")
    df.to_csv(f"../data/raw/{name}.csv")
    print(f"{name} downloaded")
```
