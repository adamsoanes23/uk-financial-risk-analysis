```python
import yfinance as yf
import pandas as pd
import os

#Create directory if it doesn't exist
os.makedirs("../data/raw", exist_ok=True)

# Define the tickers and their corresponding filenames
tickers = {
    "FTSE100": "^FTSE",
    "UK_Gov_Bonds": "ISF.L",
    "GBPUSD": "GBPUSD=X",
    "SP500": "^GSPC"
}

# Download and save each dataset
for name, ticker in tickers.items():
    df = yf.download(ticker, start="2015-01-01", end="2025-01-01")
    df.to_csv(f"../data/raw/{name}.csv")
    print(f"{name} downloaded")
```
